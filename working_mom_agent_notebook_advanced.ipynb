```
import json
from datetime import datetime

# -------------------------
# AGENT 1: DAILY TASK PLANNER
# -------------------------
def plan_day(name, wake_time="6:30 AM", work_blocks=2):
    tasks = [
        {"time": wake_time, "task": "Wake up & hydrate"},
        {"time": "7:00 AM", "task": "Prepare breakfast + get child ready"},
    ]
    if work_blocks >= 1:
        tasks.append({"time": "9:00 AM", "task": "Focused Work Block 1"})
    tasks.append({"time": '1:00 PM', "task": "Lunch + 15 min walk"})
    if work_blocks >= 2:
        tasks.append({"time": "3:00 PM", "task": "Focused Work Block 2"})
    tasks.extend([
        {"time": "7:00 PM", "task": "Family/play time"},
        {"time": "9:30 PM", "task": "Relax, reduce screen time"},
        {"time": "10:00 PM", "task": "Sleep"}
    ])
    return {"name": name, "tasks": tasks}


# -------------------------
# AGENT 2: NUTRITION ADVISOR
# -------------------------
def nutrition_plan():
    return {
        "breakfast": ["Idli + tomato chutney", "Muesli + milk", "Moong dal sprouts"],
        "lunch": ["Rice + sambar + greens poriyal", "Chapati + dal + veg curry"],
        "snacks": ["Fruit bowl (apple/guava)", "Roasted channa"],
        "dinner": ["Chapati + kurma", "Veg upma", "Light dosa + chutney"]
    }


# -------------------------
# AGENT 3: CHILD ROUTINE HELP
# -------------------------
def child_routine():
    return {
        "wake_time": "7:30 AM",
        "nap_time": "1:30 PM â€“ 3:00 PM",
        "play_time": "6:00 PM â€“ 7:00 PM",
        "sleep_time": "9:00 PM"
    }


# -------------------------
# AGENT 4: STRESS REDUCTION COACH
# -------------------------
def stress_relief():
    return [
        "Take 5 deep breaths before starting work",
        "10-minute evening walk",
        "Limit social media scrolling",
        "Drink water every 2 hours"
    ]


# -------------------------
# MASTER AGENT
# -------------------------
def master_agent(name="User"):
    return {
        "daily_plan": plan_day(name),
        "nutrition": nutrition_plan(),
        "child_routine": child_routine(),
        "stress_relief": stress_relief()
    }


# Generate output
output = master_agent("Vaishnavi")
output
```

```
# Export output for Kaggle submission
with open('/mnt/data/advanced_agent_output.json', 'w') as f:
    json.dump(output, f, indent=4)

'/mnt/data/advanced_agent_output.json saved successfully.'
```
